```python
import torch

# 1. 加载.pt文件（替换为你的文件路径）
file_path = "/home/bohao/LuvRobot/Y1_1Pace/data/Y1_1_sim/chrip_data.pt"
data = torch.load(file_path)
```

```python
# 2. 第一步：查看数据的类型和基本信息（关键）
print("=== 数据基本信息 ===")
print(f"数据类型: {type(data)}")  # 常见类型：torch.Tensor、dict、list、tuple等
```

```python
# 3. 分类型查看详细内容
if isinstance(data, torch.Tensor):
    # 场景1：数据是张量（最常见）
    print(f"张量形状: {data.shape}")
    print(f"张量数据类型: {data.dtype}")
    print(f"张量设备: {data.device}")  # 查看数据在CPU/GPU
    print(f"前5行前5列数据:\n{data[:5, :5]}")  # 预览部分数据
    print(f"数据范围: 最小值={data.min().item()}, 最大值={data.max().item()}")

elif isinstance(data, dict):
    # 场景2：数据是字典（常用，存储多组数据）
    print(f"字典的键: {list(data.keys())}")
    # 遍历每个键的值
    for key, value in data.items():
        print(f"\n--- 键: {key} ---")
        print(f"值类型: {type(value)}")
        if isinstance(value, torch.Tensor):
            print(f"值形状: {value.shape}")
            print(f"值示例:\n{value[:3]}")  # 预览前3个元素

elif isinstance(data, list) or isinstance(data, tuple):
    # 场景3：数据是列表/元组
    print(f"列表/元组长度: {len(data)}")
    # 查看前3个元素
    for i, item in enumerate(data[:3]):
        print(f"\n--- 第{i}个元素 ---")
        print(f"类型: {type(item)}")
        if isinstance(item, torch.Tensor):
            print(f"形状: {item.shape}")
            print(f"示例:\n{item[:2]}")
```
